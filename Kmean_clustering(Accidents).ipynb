```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Enable inline plotting for notebooks
%matplotlib inline

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Data_set/ML project/updated_with_coordinates.csv")

# Data Cleaning
# Handle missing values
df = df.dropna(subset=['Latitude', 'Longitude'])  # Drop rows where latitude/longitude is missing
df = df.fillna(0)  # Fill other missing numerical values with 0

# Convert 'Incident Date' to datetime objects
df['Incident Date'] = pd.to_datetime(df['Incident Date'], errors='coerce')

# Extract date-related features
df['Year'] = df['Incident Date'].dt.year
df['Month'] = df['Incident Date'].dt.month
df['Day'] = df['Incident Date'].dt.day
df['DayOfWeek'] = df['Incident Date'].dt.dayofweek

# Drop the original date column
df = df.drop(['Incident Date'], axis=1)

# Define features for clustering
features_for_clustering = [
    'Victims Killed', 'Victims Injured', 'Suspects Killed',
    'Suspects Injured', 'Suspects Arrested', 'Latitude', 'Longitude',
    'Year', 'Month', 'Day', 'DayOfWeek'
]

# Outlier Detection and Removal using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

for column in features_for_clustering:
    df = remove_outliers_iqr(df, column)

# Feature Scaling (MinMaxScaler)
scaler = MinMaxScaler()
df[features_for_clustering] = scaler.fit_transform(df[features_for_clustering])

# Handle Categorical Data - Label Encoding
categorical_features = ['State', 'City Or County', 'Address']
for feature in categorical_features:
    if feature in df.columns:
        df[feature] = df[feature].astype(str).fillna('Unknown')  # Convert to string and handle NaN
        label_encoder = LabelEncoder()
        df[feature] = label_encoder.fit_transform(df[feature])

# Add back categorical features to the feature list
features_for_clustering.extend(categorical_features)

# Clustering with different K values
inertia_values = []
k_values = range(2, 11)  # Trying k from 2 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    df['Cluster'] = kmeans.fit_predict(df[features_for_clustering])  # Use all preprocessed data
    inertia_values.append(kmeans.inertia_)

# Plotting the Elbow method
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()

# Display cluster information for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    df['Cluster'] = kmeans.fit_predict(df[features_for_clustering])

    print(f"\n--- Cluster Analysis for k = {k} ---")
    print(df['Cluster'].value_counts())  # Show number of points per cluster

    # Calculate mean values for numerical columns by cluster
    cluster_means = df.groupby('Cluster')[features_for_clustering].mean()
    print("\nCluster Means:")
    print(cluster_means)

    # Visualize clusters with a scatter plot (using 2 features for easier viewing)
    plt.figure(figsize=(8, 6))
    if 'Latitude' in df.columns and 'Longitude' in df.columns:
        plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis', s=50, alpha=0.7)
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
    elif len(features_for_clustering) >= 2:
        plt.scatter(df[features_for_clustering[0]], df[features_for_clustering[1]], c=df['Cluster'], cmap='viridis', s=50, alpha=0.7)
        plt.xlabel(features_for_clustering[0])
        plt.ylabel(features_for_clustering[1])
    else:
        print("Not enough columns to create a scatter plot.")

    plt.title(f'Cluster Visualization (k={k})')
    plt.show()
```
